```python
#Initilise Liberiaes
import pandas as pd
from sklearn.preprocessing import LabelEncoder
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
#Getting data from CSV File
Data_Raw = pd.read_csv("/content/drive/MyDrive/PGD/house_price.csv.csv")
Data_Raw.head(10)
```

```python
#getting basic insight about the raw data
Data_Raw.info()
```

```python
#Analysing the missing values in dataset
Percent_Missing = Data_Raw.isnull().sum() * 100 / len(Data_Raw)
Percent_Missing
```

1.   'society' have 40% of missing data, the column can be droped.
2.   'size, and  'bath'are having minmal number of missing data (< 1%) same can be removed.
3.   'balcony' have 4% missing data, however there might be some relationship between size and balcony. Corelation of the same to be checked.









```python
#'society' have 40% of missing data, the column can be droped.
Data_EDA = Data_Raw.drop(['society'],axis=1)
# 'total_sqft' and 'sqft' are same information, let drop 'total_sqft'
Data_EDA.drop(['total_sqft'],axis=1, inplace=True)
# The data in 'location' also have lot of garbage value, this also can be dropeed
Data_EDA.drop(['location'],axis=1, inplace=True)
Data_EDA.info()
Data_EDA.head(10)
```

```python
#'size' and 'bath' are having minmal number of missing data (< 1%) same can be removed.
Data_EDA.dropna(subset=['size','bath'],inplace=True)
Data_EDA.info()
```

```python
#'balcony' have 4% missing data, however there might be some relationship between 'size' and 'balcony'. Corelation of the same to be checked.
#Anayzing the unique values in 'size' coloumn
Data_EDA['size'].unique()
```

```python
# 'balcony' have 4% missing data
# It can replace with "0" assuming the user forgot to enter the data
Data_EDA['balcony'].fillna(0, inplace = True)
Data_EDA.info()
```

```python
# Converting the 'possession date' to a numeric data
Data_EDA['possession date'] = Data_EDA['possession date'].replace('Ready to move', '0')
Data_EDA['possession date'] = Data_EDA['possession date'].astype(int)
Data_EDA['possession date'] = Data_EDA['possession date'] / 365 + 1900
Data_EDA['possession date'] = Data_EDA['possession date'].astype(int)
Data_EDA['possession date'].unique()
```

```python
# Adding Label conding to 'possession date'
le = LabelEncoder()
Data_EDA['possession date'] = le.fit_transform(Data_EDA['possession date'])
Data_EDA['possession date'].unique()
```

```python
# Converting the 'size' to numeric data
Data_EDA['size'] = Data_EDA['size'].str.extract('(\d+)').astype(float)
Data_EDA.head()
```

```python
# Adding one hot encoding to 'area type'
Data_EDA = pd.concat([Data_EDA, pd.get_dummies(Data_EDA['area_type'], prefix='area_type')], axis=1)
Data_EDA.drop(['area_type'],axis=1, inplace=True)
Data_EDA.info()
```

```python
# Building a new variable called "price per Sqft"
Data_EDA['price per sqft'] = Data_EDA['price'] / Data_EDA['sqft']
Data_EDA.info()
```
